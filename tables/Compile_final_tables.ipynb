# Assembly statistics for assembly groups

# Eukaryotic MAG summary table

```python
mag_quast = pd.read_csv('mag_quast_report.txt', sep='\t', index_col=0)
mag_quast.index = mag_quast.index.str.replace('_', '-').str.replace('-bin', '_bin')
```

```python
pfam = pd.read_csv('pfam-annotation.csv',index_col=0)
kegg = pd.read_csv('total_ko-annot.tab', sep=' ', index_col=0, header=None)
kegg.index = kegg.index.str.strip('.kofamscan.mapper.out')
kegg.columns = ['total_kegg']
total_proteins = pd.read_csv('total_proteins.tab', sep=' ', index_col=0)
total_proteins.index = total_proteins.index.str.strip('.faa')
unique = pd.read_csv('uniqe_pfam_kegg.csv',index_col=0)
eukcc = pd.read_csv('eukcc_completeness.csv',index_col=0)
busco = pd.read_csv('busco-update-2021-01-12.tsv',sep='\t',index_col=0)
busco_cc = pd.read_csv('EUK_BUSCO_CC.csv', index_col=0)
tax = pd.read_csv('2021-03-final-estimated_taxonomy.csv', index_col=0)
tax.columns = ['eukulele_taxonomy']
tax_split = pd.read_csv('2021-03-final-estimated_taxonomy_split.csv', index_col=0)
mmseqs_tax = pd.read_csv('../data/mmseqs-taxonomy.tsv', sep='\t', index_col=0)
mmseqs_tax['combo']=mmseqs_tax['kingdom'].str.cat(mmseqs_tax[['clade', 'phylum', 'class', 'subclass', 'order', 'family',
       'species',]].astype(str), sep=";").str.replace('.','-')
mmseqs_tax['mmseqs_taxonomy'] = 'Eukaryota;' + mmseqs_tax['combo'].astype(str)
coding = pd.read_csv('mag-coding-density.csv', index_col=0)
coding.columns=['proportion_coding_sequence_mean', 'proportion_repeat_mean', 'proportion_coding_sequence_std', 'proportion_repeat_std']
```

```python
full_stats = mag_quast.merge(total_proteins, left_index=True, right_index=True).merge(pfam, 
        left_index=True, right_index=True).merge(kegg, left_index=True, right_index=True).merge(unique,
        left_index=True, right_index=True).merge(busco_cc, 
        left_index=True, right_index=True).merge(eukcc, 
        left_index=True, right_index=True, how='outer').merge(tax, 
        left_index=True, right_index=True).merge(tax_split, left_index=True, right_index=True).merge(mmseqs_tax, left_index=True, right_index= True)
subset_for_table = full_stats[['eukulele_taxonomy','mmseqs_taxonomy','Total length','# contigs (>= 0 bp)', 
       'Total length (>= 5000 bp)', 'Total length (>= 10000 bp)',
         'Total length (>= 50000 bp)', 
        'GC (%)', 'N50', 'L50', '0', 'num_annotated', 'total_kegg', 'pfam','kegg', 'Completeness','Contamination', 'eukcc_completeness','eukcc_contamination','eukcc_lineage']]
subset_for_table.columns = ['eukulele_taxonomy','mmseqs_taxonomy','total_length', 'total_num_contigs', 'total_length_5kb',
       'total_length_10kb', 'total_length_50kb', 
       'GC_percent', 'N50', 'L50', 'num_predicted_proteins', 'num_proteins_annotated_pfam', 
        'num_proteins_annotated_kegg', 'num_unique_pfam','num_unique_kegg', 'busco_eukaryota_odb10_completeness',
        'busco_eukaryota_odb10_contamination','eukcc_completeness','eukcc_contamination','eukcc_lineage']

subset_for_table['proportion_proteins_annoated_pfam']=100*(subset_for_table.num_proteins_annotated_pfam/subset_for_table.num_predicted_proteins)
subset_for_table['proportion_proteins_annoated_kegg']=100*(subset_for_table.num_proteins_annotated_kegg/subset_for_table.num_predicted_proteins)
subset_for_table = subset_for_table[['eukulele_taxonomy','mmseqs_taxonomy','total_length', 'total_num_contigs', 'total_length_5kb',
       'total_length_10kb', 'total_length_50kb', 
       'GC_percent', 'N50', 'L50', 'num_predicted_proteins', 'num_proteins_annotated_pfam', 
                  'proportion_proteins_annoated_pfam','num_unique_pfam','num_proteins_annotated_kegg', 'proportion_proteins_annoated_kegg','num_unique_kegg', 'busco_eukaryota_odb10_completeness','busco_eukaryota_odb10_contamination',
                            'eukcc_completeness','eukcc_contamination','eukcc_lineage']]
subset_for_table['ocean_region'] = subset_for_table.index.str.split('-').str[0]
subset_for_table['depth'] = subset_for_table.index.str.split('-').str[2]
```

```python
neweuk_dict = pd.read_csv('../rename/renamed-eukaryotic-mags.tsv', sep='\t', index_col=0).to_dict()['new_mag_name']
EUK_MAGS_TO_DROP = list(pd.read_csv('MAGSTODROP.list', header=None)[0])
GOODMAGS = subset_for_table.index.difference(EUK_MAGS_TO_DROP)
final_table = subset_for_table.loc[GOODMAGS]
final_table.index = final_table.index.map(neweuk_dict)
sf_dict = {'1':'0.8-5', '2':'5-20', '3':'20-180', '4':'180-2000'}
final_table['size_fraction'] = final_table.index.str.split('_').str[1].str[-1]
final_table['size_fraction']=final_table.size_fraction.map(sf_dict)
for x in ['busco_eukaryota_odb10_completeness','busco_eukaryota_odb10_contamination', 'proportion_proteins_annoated_pfam','proportion_proteins_annoated_kegg',]:
    final_table[x]=final_table[x].round(2)

final_table = final_table.merge(coding, left_index=True, right_index=True)
```

```python
EukMAGDF = final_table.copy()
```

```python
fungi_int = EukMAGDF[EukMAGDF.mmseqs_taxonomy.str.contains('Fungi')==True].index
sar = EukMAGDF[EukMAGDF.mmseqs_taxonomy.str.contains('Eukaryota;-;Sar;-;-')==True].index
dino = EukMAGDF[EukMAGDF.mmseqs_taxonomy.str.contains('Eukaryota;-;Sar;-;Dinophyceae')==True].index
amoeba = EukMAGDF[EukMAGDF.mmseqs_taxonomy.str.contains('Eukaryota;-;Amoebozoa')==True].index
apuso =  EukMAGDF[EukMAGDF.mmseqs_taxonomy.str.contains('Eukaryota;-;Apusozoa')==True].index
Haptista = EukMAGDF[EukMAGDF.mmseqs_taxonomy.str.contains('Eukaryota;-;Prymnesiophyceae;Haptista')==True].index
Choa = EukMAGDF[EukMAGDF.mmseqs_taxonomy.str.contains('Choanoflagellata')==True].index
Crypto = EukMAGDF[EukMAGDF.mmseqs_taxonomy.str.contains('Cryptophyceae')==True].index



EukMAGDF['groups']=EukMAGDF.eukulele_taxonomy.str.split(';').str[1:3].str.join(';')
EukMAGDF['groups'].unique()
EukMAGDF.loc[fungi_int, 'groups']='Fungi'
EukMAGDF.loc[sar, 'groups'] = 'SAR'
EukMAGDF.loc[dino, 'groups'] = 'Dinoflagellata'
EukMAGDF.loc[amoeba, 'groups'] = 'Amoebozoa'
EukMAGDF.loc[apuso, 'groups'] = 'Apusozoa'
EukMAGDF.loc[Haptista, 'groups'] = 'Haptophyta'
EukMAGDF.loc[Choa, 'groups'] = 'Choanozoa'
EukMAGDF.loc[Crypto, 'groups'] = 'Cryptophyta'

EukMAGDF['groups_full'] =EukMAGDF.groups
EukMAGDF.loc[EukMAGDF['groups']=='Metazoa;Arthropoda', 'groups']='Metazoa'
EukMAGDF.loc[EukMAGDF['groups']=='Metazoa;Cnidaria', 'groups']='Metazoa'
EukMAGDF.loc[EukMAGDF['groups']=='Metazoa;-', 'groups']='Metazoa'
EukMAGDF.loc[EukMAGDF['groups']=='Metazoa;Chordata', 'groups']='Metazoa'
EukMAGDF.loc[EukMAGDF['groups']=='Metazoa;Mollusca', 'groups']='Metazoa'
EukMAGDF.loc[EukMAGDF['groups']=='Archaeplastida;Chlorophyta', 'groups']='Chlorophyta'
EukMAGDF.loc[EukMAGDF['groups']=='Alveolata;Ciliophora', 'groups']='SAR'
EukMAGDF.loc[EukMAGDF['groups']=='Alveolata;Dinoflagellata', 'groups']='Dinoflagellata'
EukMAGDF.loc[EukMAGDF['groups']=='Alveolata;Ciliophora', 'groups']='SAR'
EukMAGDF.loc[EukMAGDF['groups']=='Stramenopiles;-', 'groups']='SAR'
EukMAGDF.loc[EukMAGDF['groups']=='Amoebozoa;-', 'groups']='Amoebozoa'
EukMAGDF.loc[EukMAGDF['groups']=='-;-', 'groups']='Eukaryota'
EukMAGDF.loc[EukMAGDF['groups']=='Hacrobia;Haptophyta', 'groups']='Haptophyta'
EukMAGDF.loc[EukMAGDF['groups']=='Hacrobia;Cryptophyta', 'groups']='Cryptophyta'
EukMAGDF.loc[EukMAGDF['groups']=='Stramenopiles;Ochrophyta', 'groups']='Ochrophyta'
```

```python
EukMAGDF = EukMAGDF[['eukulele_taxonomy', 'mmseqs_taxonomy', 'groups', 'total_length',
       'total_num_contigs', 'total_length_5kb', 'total_length_10kb',
       'total_length_50kb', 'GC_percent', 'N50', 'L50',
       'proportion_coding_sequence_mean', 'proportion_coding_sequence_std', 
       'proportion_repeat_mean', 'proportion_repeat_std',
       'num_predicted_proteins', 'num_proteins_annotated_pfam',
       'proportion_proteins_annoated_pfam', 'num_unique_pfam',
       'num_proteins_annotated_kegg', 'proportion_proteins_annoated_kegg',
       'num_unique_kegg', 'busco_eukaryota_odb10_completeness',
       'busco_eukaryota_odb10_contamination', 'eukcc_completeness',
       'eukcc_contamination', 'eukcc_lineage', 'ocean_region', 'depth',
       'size_fraction', ]]
EukMAGDF
```

```python
EukMAGDF.to_csv('final/TableS2_EukaryoticMAG.csv')
```
